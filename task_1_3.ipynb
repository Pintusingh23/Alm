## 1. Main Baseline

The primary baseline the paper compares against is the **Uniform Combination solution** (referred to as `unif` in the paper's tables). In this baseline, all $p$ base kernels are combined with equal weights: $K_{unif} = \frac{1}{p} \sum_{k=1}^{p} K_k$, and a standard SVM or KRR is trained on this combined kernel without any learning of weights.

The paper also compares against other existing learning kernel methods such as **L1-SVM** (Lanckriet et al., 2004) and **L2-KRR** (Cortes et al., 2009), but the uniform combination is positioned as the key benchmark because, as the authors note, no prior learning kernel algorithm had been able to consistently beat it.

## 2. Limitation of the Baseline

The paper identifies that the **uniform combination ignores the relative quality of individual kernels** with respect to the classification task at hand. By assigning equal weight to every base kernel regardless of how informative it is, the uniform solution dilutes the contribution of highly aligned kernels with the noise from poorly aligned ones. The authors note in Section 1 that this limitation had been widely observed in the community — multiple NIPS workshops highlighted that no learning kernel algorithm had consistently improved over uniform combination, suggesting the baseline was strong but also that prior learning approaches were fundamentally flawed in how they measured kernel quality.

## 3. How the Proposed Method Overcomes This Limitation

The paper overcomes this by replacing the implicit equal-weighting of kernels with an explicit, theoretically grounded measure of each kernel's usefulness — **centered alignment** with the target label kernel $K_y = yy^\top$. By centering both the base kernels and the target kernel before computing alignment, the method removes the trivial constant component that caused uncentered alignment to be an unreliable measure. ALIGNF then maximizes this centered alignment jointly over all kernel weights via a QP, finding the combination that best reflects the label structure in the training data.

## 4. One Condition Under Which the Paper's Method Would NOT Outperform the Baseline

The paper's method would likely **not outperform the uniform combination when all base kernels have nearly equal centered alignment with the target**. In this scenario, the QP solved by ALIGNF would produce weights close to uniform anyway — but the overhead of computing centered kernel matrices, solving the QP, and then training SVM on the combined kernel adds computational cost with no quality benefit.

More importantly, this situation reveals a deeper issue: if the base kernels are themselves poorly chosen (e.g., all are Gaussian RBF kernels with only slightly different bandwidths applied to an irrelevant feature set), then no reweighting — uniform or learned — can extract meaningful signal. The ALIGNF method can only redistribute weight among existing base kernels; it cannot escape a poor base kernel family. In such cases, the uniform combination is computationally cheaper and equally ineffective, meaning the method's advantage disappears entirely.

This failure condition is my own reasoned inference: the paper's empirical results (Table 1, 2, 3) show consistent improvements, but all experiments use diverse base kernel families (polynomial + RBF + linear). A setting with a homogeneous, poorly-constructed kernel family is not tested and would expose the boundary of the method's advantage.
